#### 2.1.4.3 了解特征

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math


# 检查每个表格的属性
def describe_and_visualize(df, table_name):
    print("\n[log] -------------------- \n")
    print(f"分析表格：{table_name}")
    print(f"表格的形状：{df.shape}")

    categorical_columns = []  # 类别型字段
    numerical_columns = []  # 数值型字段

    # 分类字段与数值字段区分
    for column in df.columns:
        if df[column].dtype == 'object' or df[column].nunique() < 30:  # 类别型或唯一值较少
            categorical_columns.append(column)
        else:
            numerical_columns.append(column)

    # 输出类别型字段的信息
    print(f"\n>>> 类别型字段（{len(categorical_columns)} 个）：{categorical_columns}")
    for column in categorical_columns:
        print(f"字段：{column}")
        print(f"{column}是类别型数据，共有{df[column].nunique()}个不同的值")
        print(df[column].value_counts())

    # 绘制类别型字段的条形图
    if categorical_columns:
        print("\n[log] 正在绘制类别型字段的统计图...")
        rows = math.ceil(len(categorical_columns) / 4)
        fig, axes = plt.subplots(rows, min(len(categorical_columns), 4), figsize=(20, 5 * rows))
        axes = axes.flatten()  # 展平方便处理
        for i, column in enumerate(categorical_columns):
            df[column].value_counts().plot(kind='bar', ax=axes[i], title=f'{column} - Count Values')
            axes[i].set_xlabel(column)
            axes[i].set_ylabel('Frequency')
        # 隐藏多余的子图
        for j in range(len(categorical_columns), len(axes)):
            axes[j].axis('off')
        plt.tight_layout()
        plt.show()

    # 输出数值型字段的信息
    print(f"\n>>> 数值型字段（{len(numerical_columns)} 个）：{numerical_columns}")
    for column in numerical_columns:
        print(f"\n字段：{column}")
        print(f"{column}是数值型数据，共有{df[column].nunique()}个不同的值")
        print(df[column].describe())

    # 绘制数值型字段的分布图
    if numerical_columns:
        print("\n[log] 正在绘制数值型字段的分布图...")
        rows = math.ceil(len(numerical_columns) / 4)
        fig, axes = plt.subplots(rows, min(len(numerical_columns), 4), figsize=(20, 5 * rows))
        axes = axes.flatten()  # 展平方便处理
        for i, column in enumerate(numerical_columns):
            sns.histplot(df[column], kde=True, bins=30, ax=axes[i])
            axes[i].set_title(f'{column} - Distribution')
            axes[i].set_xlabel(column)
            axes[i].set_ylabel('Frequency')
        # 隐藏多余的子图
        for j in range(len(numerical_columns), len(axes)):
            axes[j].axis('off')
        plt.tight_layout()
        plt.show()


# 对每个表格调用函数
describe_and_visualize(ad_feature, "ad_feature")
describe_and_visualize(raw_sample, "raw_sample")
describe_and_visualize(user_profile, "user_profile")
describe_and_visualize(behavior_log, "behavior_log")

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# 定义一个函数来绘制子图中的热力图
def plot_heatmap_subplot(ax, df, title):
    correlation_matrix = df.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar_kws={"shrink": .5},
                vmin=-1, vmax=1, ax=ax)
    ax.set_title(title)
    # x轴标签旋转
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)


# 为 behavior_log 临时编码
behavior_log_encoded = behavior_log.copy()
# 将 btag 列映射为数值
btag_mapping = {'pv': 1, 'cart': 2, 'fav': 3, 'buy': 4}
behavior_log_encoded['btag'] = behavior_log_encoded['btag'].map(btag_mapping)

# 创建 2x2 子图
fig, axes = plt.subplots(2, 2, figsize=(16, 12))  # 调整图形大小

# 绘制每个数据集的热力图
plot_heatmap_subplot(axes[0, 0], user_profile, 'User Profile Correlation Matrix')
plot_heatmap_subplot(axes[0, 1], raw_sample, 'Raw Sample Correlation Matrix')
plot_heatmap_subplot(axes[1, 0], behavior_log_encoded, 'Behavior Log Correlation Matrix')
plot_heatmap_subplot(axes[1, 1], ad_feature, 'Ad Feature Correlation Matrix')

# 调整布局
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 1. 合并 raw_sample 和 user_profile。通过主键user_id 字段进行连接。
merged_data = pd.merge(raw_sample, user_profile, on='user_id', how='left')

# 2. 再将上述结果与 ad_feature 合并。通过主键adgroup_id 字段进行连接。
ad_u_data = pd.merge(merged_data, ad_feature, on='adgroup_id', how='left')

# 使用左连接确保所有 raw_sample 数据保留，即使对应 user_profile 或 ad_feature 中没有匹配项。
# how='left' 是保留左表所有数据，并将右表匹配的部分合并过来，如果没有匹配项则用 NaN 填充右表的缺失值。
# 相当于：
# SELECT * FROM raw_sample AS rs 
# LEFT JOIN user_profile AS up ON rs.user_id = up.user_id
# LEFT JOIN ad_feature AS af ON rs.adgroup_id = af.adgroup_id;


# 查看合并后的数据
print(f"最终数据形状: {ad_u_data.shape}")
print(ad_u_data.head())  # (7114606, 21)

# 将合并后的数据保存
ad_u_data.to_csv(f'{processed_data_path}/ad_u_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


def pareto_analysis(data, group_col, value_col, title):
    # 统计各组的总量
    grouped = data.groupby(group_col)[value_col].sum().reset_index()
    grouped = grouped.sort_values(by=value_col, ascending=False)

    # 计算累计百分比
    grouped['cum_value'] = grouped[value_col].cumsum()
    grouped['cum_percentage'] = grouped['cum_value'] / grouped[value_col].sum()
    grouped['cum_percentage_index'] = np.arange(1, len(grouped) + 1) / len(grouped)

    # 绘制累计百分比曲线
    plt.figure(figsize=(10, 6))
    plt.plot(grouped['cum_percentage_index'], grouped['cum_percentage'], label='Cumulative Percentage')
    plt.axhline(y=0.8, color='r', linestyle='--', label='80% Threshold')
    plt.axvline(x=0.2, color='g', linestyle='--', label='20% Threshold')
    plt.title(title)
    plt.xlabel('Cumulative Percentage of Groups')
    plt.ylabel('Cumulative Percentage of Values')
    plt.legend()
    plt.grid()
    plt.show()


# 分析品牌广告量
pareto_analysis(ad_u_data, group_col='brand', value_col='adgroup_id', title='Pareto Analysis: Brand vs Ad Count')

# 分析用户点击量
pareto_analysis(ad_u_data, group_col='user_id', value_col='clk', title='Pareto Analysis: User vs Clicks')

# 分析用户未点击量
pareto_analysis(ad_u_data, group_col='user_id', value_col='nonclk', title='Pareto Analysis: User vs Non-Clicks')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


def plot_density_with_mean(data, feature, price_col='price', clk_col='clk', nonclk_col='nonclk'):
    unique_values = data[feature].dropna().unique()  # 获取属性的唯一值
    n_unique = len(unique_values)  # 唯一值数量

    fig, axes = plt.subplots(1, n_unique, figsize=(8 * n_unique, 6), sharey=True)
    if n_unique == 1:  # 如果只有一个唯一值（特殊情况）
        axes = [axes]

    for ax, value in zip(axes, unique_values):
        # 筛选出当前分组数据
        subset = data[data[feature] == value]

        # 绘制点击（clk）和不点击（nonclk）的概率密度
        sns.kdeplot(subset[price_col][subset[clk_col] == 1], ax=ax, color='red', label='clk', fill=True)
        sns.kdeplot(subset[price_col][subset[nonclk_col] == 1], ax=ax, color='green', label='nonclk', fill=True)

        # 计算均值
        clk_mean = subset[price_col][subset[clk_col] == 1].mean()
        nonclk_mean = subset[price_col][subset[nonclk_col] == 1].mean()

        # 绘制竖虚线表示均值
        ax.axvline(clk_mean, color='red', linestyle='--', label=f'clk mean: {clk_mean:.2f}')
        ax.axvline(nonclk_mean, color='green', linestyle='--', label=f'nonclk mean: {nonclk_mean:.2f}')

        # 设置标题和图例
        ax.set_title(f"{feature} = {value}")
        ax.set_xlabel("Price")
        ax.set_ylabel("Density")
        ax.legend()

    plt.tight_layout()
    plt.show()


# 创建ad_u_data的副本并排除price>2000的元组，这样可视化效果更好（不会受极端值影响）
ad_u_data_filtered = ad_u_data[ad_u_data['price'] <= 2000].copy()

# 按性别绘制（final_gender_code）
plot_density_with_mean(ad_u_data_filtered, feature='final_gender_code')

# 按年龄分层绘制（age_level）
plot_density_with_mean(ad_u_data_filtered, feature='age_level')

# 按消费档次绘制（pvalue_level）
plot_density_with_mean(ad_u_data_filtered, feature='pvalue_level')

# 按购物深度绘制（shopping_level）
plot_density_with_mean(ad_u_data_filtered, feature='shopping_level')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
processed_data_path = "G:/DataSets/Ali_Display_Ad_Click/processed_data"
ad_u_data = pd.read_csv(f'{processed_data_path}/ad_u_data.csv')

# 按广告组进行聚合
data_adgroup = ad_u_data.groupby('adgroup_id')['clk'].agg(
    展示量='count',
    点击量='sum',
    点击率='mean'
)

# 定义点击量区间(因为总共就391786个clk，所以可以确信最大值<391786)
bins = [0, 1, 3, 5, 10, 20, 50, 100, 391786]
labels = ['0', '1-2', '3-4', '5-9', '10-19', '20-49', '50-99', '>=100']

data_adgroup['点击量区间'] = pd.cut(data_adgroup['点击量'], bins=bins, labels=labels, right=False)

# 按点击量区间进行聚合
data_adgroup_clk = data_adgroup.groupby('点击量区间').agg({
    '点击量': 'sum',
    '展示量': 'sum'
})

# 计算点击率
data_adgroup_clk['点击率'] = data_adgroup_clk['点击量'] / data_adgroup_clk['展示量'] * 100

# 提取数据
x = data_adgroup_clk.index
y1 = data_adgroup_clk['展示量'] - data_adgroup_clk['点击量']  # 无点击量
y2 = data_adgroup_clk['点击量']  # 点击量
y3 = data_adgroup_clk['点击率']  # 点击率

plt.rcParams['font.sans-serif'] = ['SimSun']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 为了显示负号
# 绘制图像
fig, ax1 = plt.subplots(figsize=(10, 6))

# 堆叠柱状图
bar1 = ax1.bar(x, y1, label='无点击量', color='lightcoral', width=0.6)
bar2 = ax1.bar(x, y2, bottom=y1, label='点击量', color='brown', width=0.6)
ax1.set_xlabel('点击量区间')
ax1.set_ylabel('展示量')
ax1.legend(loc='upper left')
ax1.tick_params(axis='y', labelcolor='brown')

# 折线图
ax2 = ax1.twinx()
ax2.plot(x, y3, color='skyblue', marker='o', label='点击率')
ax2.set_ylabel('点击率 (%)')
ax2.tick_params(axis='y', labelcolor='skyblue')

# 显示点击率的数值
for i, v in enumerate(y3):
    ax2.text(i, v + 0.2, f'{v:.2f}%', ha='center', va='bottom', fontsize=10, color='blue')

# 标题和整体布局
fig.suptitle('各点击数区间的广告点击率情况', fontsize=16)
fig.tight_layout()

plt.show()
```
